```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('C:\\Users\\Yaswanth\\Desktop\\DSP\\realestate.csv')
df.head()
```

```python
df1 = df.drop(columns = ['Suburb','Address','SellerG','Postcode','Date','Car','YearBuilt','CouncilArea','Lattitude','Longtitude'])
df1.head()
```

```python
df_Clean = df1.dropna()
df_Clean.head()
```

```python
df_Clean.isnull().sum()
```

```python
df_Clean.head(2)
```

```python
LabelEncoder = LabelEncoder()
df_Clean.iloc[:,1] = LabelEncoder.fit_transform(df_Clean.iloc[:,1])
df_Clean.iloc[:,3] = LabelEncoder.fit_transform(df_Clean.iloc[:,3])
#df_Clean.loc[:,'Type'] = LabelEncoder.fit_transform(df_Clean.loc[:,'Type'])
#df_Clean.loc[:,'Method'] = LabelEncoder.fit_transform(df_Clean.loc[:,'Method'])
```

```python
df_Clean.head()
```

```python
df_Clean.corr()
```

```python
df_cor = df_Clean.corr()
corr = df_cor[df_cor['Price']>0.3]
corr
```

```python
 #from sklearn.linear_model import LinearRegression
x = np.array(df_Clean['Rooms']).reshape(-1,1)
y = np.array(df_Clean['Price'])
reg = LinearRegression().fit(x,y)
reg
```

```python
print('Coefficient of determinatin = ', reg.score(x,y))
print('Intercept = ',reg.intercept_)
print('Slope = ', reg.coef_)
```

```python
x1 = np.array(df_Clean.loc[:,['Rooms','Bedroom2','Bathroom']]).reshape(-1,3)
y1 = np.array(df_Clean['Price'])
reg1 = LinearRegression().fit(x1,y1)
reg1
```

```python
print('Coefficient of determinatin = ', reg1.score(x1,y1))
print('Intercept = ',reg1.intercept_)
print('Slope = ', reg1.coef_)
```

```python
y_pred = reg1.predict(x1)
y_pred
```

```python
df_Clean['Predicted_price'] = (y_pred)
```
